```python
import pandas as pd
import numpy as np
import re
import torch
import spacy
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from transformers import BertTokenizer, BertModel
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from imblearn.over_sampling import SMOTE
```

```python
# Load Data
train_data = pd.read_csv('synthetic_data.csv')
train_data.drop_duplicates(inplace=True)
test_data = pd.read_csv('real_data.csv', nrows=1000)
```

```python
print(train_data.shape)
train_data['label'].value_counts().plot(kind='bar')
```

```python
print(test_data.shape)
test_data['label'].value_counts().plot(kind='bar')
```

```python
# Preprocessing Function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www.\S+', '', text)  # Remove URLs
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)  # Remove special characters
    text = ' '.join([word for word in text.split() if word not in stopwords.words('english')])  # Remove stopwords
    return text

train_data['text'] = train_data['text'].apply(preprocess_text)
test_data['text'] = test_data['tweet'].apply(preprocess_text)
```

```python
# Convert text to embeddings using BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=50)
    with torch.no_grad():
        outputs = bert_model(**inputs)
    return outputs.last_hidden_state[:, 0, :].squeeze().numpy()
```

```python
train_data['embedding'] = train_data['text'].apply(get_bert_embedding)
test_data['embedding'] = test_data['text'].apply(get_bert_embedding)
```

```python
# Convert embeddings to numpy arrays
X_train = np.stack(train_data['embedding'].values)
X_test = np.stack(test_data['embedding'].values)
y_train = train_data['label']
y_test = test_data['label']
```

```python
# Handle class imbalance using SMOTE
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Train Classifier with Hyperparameter Tuning
classifier = RandomForestClassifier()
param_grid = {'n_estimators': [50, 100, 200], 'max_depth': [10, 20, None]}
grid_search = GridSearchCV(classifier, param_grid, scoring='f1', cv=3)
grid_search.fit(X_train_resampled, y_train_resampled)
```

```python
# Best model
best_model = grid_search.best_estimator_

# Evaluate on real data
y_pred = best_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Plot Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
